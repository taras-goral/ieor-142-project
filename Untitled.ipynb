```python
bart_files = os.listdir("C:\\Users\\Taras\\Documents\\Jupyter\\ieor_142\\project\\bart data")
bart = pd.DataFrame()
os.chdir("C:\\Users\\Taras\\Documents\\Jupyter\\ieor_142\\project\\bart data")

bart_2022 = pd.read_csv(r"C:\Users\Taras\Documents\Jupyter\ieor_142\project\bart data\date-hour-soo-dest-2022.csv", header=None)

sf = ['16TH', '24TH', 'BALB', 'CIVC', 'COLM', 'DALY', 'EMBR', 'GLEN', 'MLBR', 'MONT', 'POWL', 'SBRN', 'SFIA', 'SSAN']
e_bay = list(set(bart_2022.iloc[:,2].unique()) - set(sf))

transbay_trips = pd.DataFrame(list(itertools.chain(itertools.product(sf, e_bay), itertools.product(e_bay, sf))))
transbay_trips['trips'] = transbay_trips[0] + transbay_trips[1]
transbay_trips = np.array(transbay_trips['trips'])

year = 2011
for file in bart_files:
    bart_iter = pd.read_csv(file, header=None)
    bart_iter["trip"] = bart_iter[2] + bart_iter[3]
    bart_iter = bart_iter.drop(columns=[2,3])
    bart_transbay = bart_iter.loc[bart_iter['trip'].isin(transbay_trips), :]
    bart_transbay = bart_transbay.drop(columns=['trip'])
    bart_transbay = bart_transbay.rename(columns={0: 'date', 1: 'hour', 4: 'trips'})
    bart_transbay = bart_transbay.groupby(['date', 'hour']).sum().reset_index()
    bart_transbay.insert(0, 'datetime', pd.to_datetime(bart_transbay["date"]) + bart_transbay["hour"].astype('timedelta64[h]'))
    bart_transbay.index = bart_transbay['datetime']
    bart_transbay = bart_transbay.drop(columns=['date', 'hour', 'datetime'])
    
    min_date = str(year) + '-01-01 00:00:00'
    max_date = str(year) + '-12-31 23:59:59'
    min_timestamp = dt.strptime(min_date, '%Y-%m-%d %H:%M:%S')
    max_timestamp = dt.strptime(max_date, '%Y-%m-%d %H:%M:%S')
    all_times=pd.DataFrame(pd.date_range(min_timestamp, max_timestamp, freq="1h",name="new_times"))
    all_times.index=all_times['new_times']
    all_times.drop(columns='new_times', inplace=True)
    
    bart_transbay=all_times.join(bart_transbay)
    bart_transbay = bart_transbay.fillna(0)
        
    bart = bart.append(bart_transbay)
    
    print(year)
    year = year + 1

bart
```

```python
def plot_entire_series(x, y, **kwargs):
    
    plt.figure(figsize=(16, 4))
    plt.plot(x, y, linewidth=2, color='black')
    
    for key, value in kwargs.items():
        plt.plot(x, value, linewidth=2, color=key)
        
    plt.xlabel('Date', fontsize=16)
    plt.ylabel('Rides', fontsize=16)
    plt.show()
    
    return None

def plot_last_fifty(x, y, **kwargs):
    
    plt.figure(figsize=(16, 4))
    plt.scatter(x[-50:], y[-50:], linewidth=3, color='black')
    plt.plot(x[-50:], y[-50:], linewidth=3, color='black')
    
    for key, value in kwargs.items():
        plt.plot(x[-50:], value[-50:], linewidth=3, color=key)
        
    plt.xlabel('Date', fontsize=16)
    plt.ylabel('Util', fontsize=16)
    plt.show()
    
    return None

def plot_last_five_hundred(x, y, **kwargs):
    
    plt.figure(figsize=(16, 4))
    plt.scatter(x[-500:], y[-500:], linewidth=3, color='black')
    plt.plot(x[-500:], y[-500:], linewidth=3, color='black')
    
    for key, value in kwargs.items():
        plt.plot(x[-500:], value[-500:], linewidth=3, color=key)
        
    plt.xlabel('Date', fontsize=16)
    plt.xticks(fontsize=12)
    plt.ylabel('Util', fontsize=16)
    plt.yticks(fontsize=12)
    plt.show()
    
    return None
```

```python
x = bartDataset.index.to_numpy()
y = bartDataset['transbayRides'].to_numpy()

# Utilization: transbay/total riders
plot_entire_series(x, y)
```
